 ### Data Analysis and Visualization (EDA)

#### Target Feature
Reviews or Scores

#### Problem Statement
Identify the key factors that impact customer satisfaction, measured through Scores or Reviews, and determine the attributes of TVs that contribute most to better customer feedback.

## 1. Uni-variate Analysis
Uni-variate analysis is the analysis of a single variable at a time. This can be used to summarize and find patterns in the data.

### For Continuous Variables (Numerical Data)
***Numerical Columns***: For columns with numerical data (continuous or discrete), histograms, boxplots, and density plots are commonly used.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Histogram for numerical columns

- ***Purpose***: To visualize the distribution of a numerical feature.
- ***Code Explanation***: The sns.histplot() function plots a histogram with a Kernel Density Estimate (KDE) curve overlay to give a smooth representation of the distribution.
- For Scores, it shows how ratings are distributed across the TVs. If the majority of TVs have scores above 4, the histogram will be skewed towards the higher values.

```python
numerical_cols = ["Scores", "Reviews", "Ratings", "Inches", "Selling_Price", "Original_Price", "Discount_Percentage"]

for col in numerical_cols:
    plt.figure(figsize=(4,2))
    sns.histplot(df[col], kde=True, bins=10, color='blue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

### Boxplot for numerical columns
- ***Purpose***: To visualize the spread and detect outliers.
- ***Code Explanation***: sns.boxplot() helps in showing the quartiles of the data, along with any outliers. The central line represents the median, and the box contains the interquartile range (IQR).
- For Selling_Price vs. Scores, you can see how price distribution correlates with customer ratings. If higher-priced TVs tend to have higher ratings, you’ll see a positive trend.

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x="Scores", y="Selling_Price", palette="Set2",hue = "Scores",legend = False)
plt.title("Boxplot of Selling Price vs Scores")
plt.xlabel("Scores")
plt.ylabel("Selling Price")
plt.show()
```

### KDE Plot for continuous columns
- ***Purpose***: To show the probability density function of the continuous variable.
- ***Code Explanation***: sns.kdeplot() generates a smoothed line that shows where data points are most concentrated.
- For Selling_Price, the KDE plot shows where most TVs fall in terms of price, giving an idea of the most common price range.

```python
plt.figure(figsize=(6, 4))
sns.kdeplot(df["Selling_Price"], fill=True, color="purple")
plt.title("KDE Plot of Selling Price")
plt.xlabel("Selling Price")
plt.ylabel("Density")
plt.show()
```

### Categorical Columns for Univariate Analysis
1. ****Brand****: Nominal categorical data (TV brands).
2. ****OS****: Nominal categorical data (Operating system of the TV).

### Count Plot for Categorical Values
- ***Purpose***: To visualize the count of each category in a categorical variable.
- ***Code Explanation***: sns.countplot() shows the frequency of each unique value in the column.
- For Brand, this plot shows how many TVs each brand has in the dataset. If Mi is the most common brand, it will have the tallest bar.

```python
categorical_cols = ["Brand", "OS"]

for col in categorical_cols:
    plt.figure(figsize=(12, 4))
    sns.countplot(data=df, x=col, palette="Set1",hue = col,legend = False)
    plt.title(f"Count of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=90)
    plt.show()
```

### Bar Plot for Brand vs. Scores
- ***Purpose***: To visualize the average value of a numerical feature (like Scores) for each category in a categorical variable (like Brand).
- ***Code Explanation***: sns.barplot() calculates the mean Scores for each Brand and displays it.
- For Brand vs. Scores, you can compare the average ratings for each brand. This can reveal which brand tends to get better customer ratings.

```python
plt.figure(figsize=(14, 4))
sns.barplot(data=df, x="Brand", y="Scores", palette="Set3",hue = "Brand")
plt.title("Brand vs. Scores")
plt.xlabel("Brand")
plt.ylabel("Scores")
plt.xticks(rotation=90)
plt.show()
```

### Pie Chart

```python
brand_counts = df['Brand'].value_counts()

# Create a pie chart
plt.figure(figsize=(4,4))
plt.pie(brand_counts, labels=brand_counts.index, autopct='%0.0f%%', startangle=90, colors=sns.color_palette("Set2", len(brand_counts)))
plt.title('Distribution of TV Brands')
plt.axis('equal') 
plt.show()
```

## Bi-variate and Multivariate Analysis

Bi-variate and multivariate analysis help us understand the relationships between two or more variables in the dataset.
Here, you have a few different combinations of variables that require different methods for analysis, depending on whether the variables are continuous or categorical.
1. Continuous and Categorical Variables
2. Continuous to Continuous Variables
3. Categorical and Categorical Variables

### 1. Continuous and Categorical Variables
This analysis helps us understand how continuous variables behave across different categories. You can use methods like ***GroupBy*** and ***Pivot Tables*** to explore the relationship between a continuous variable and a categorical one.
##### Example Combinations:
- Selling_Price (Continuous) vs Brand (Categorical)
- Discount Percentage (Continuous) vs OS (Categorical)

### GroupBy 
This allows us to group data by a categorical variable and calculate statistics (like mean, sum, etc.) for a continuous variable.

```python
# Group by 'Brand' and calculate the mean of 'Selling_Price'
grouped_data = df.groupby('Brand')['Selling_Price'].mean()
print(grouped_data)

# Group by 'OS' and calculate the mean of 'Discount_Percentage'
grouped_os_data = df.groupby('OS')['Discount_Percentage'].mean()
print(grouped_os_data)
```

### Pivot Tables
A pivot table allows us to summarize and aggregate data, giving a clear comparison between continuous variables for different categories.

```python
# Pivot table to get the average Selling_Price for each Brand
pivot_data = df.pivot_table(values='Selling_Price', index='Brand', aggfunc='mean')
print(pivot_data)

# Pivot table to get the average Discount Percentage for each OS
pivot_os_data = df.pivot_table(values='Discount_Percentage', index='OS', aggfunc='mean')
print(pivot_os_data)
```

### Continuous to Continuous Variables (Correlation Plot)
In this analysis, we aim to understand the relationship between two continuous variables. Correlation analysis is often used to measure the strength and direction of the relationship between continuous variables. ***Heatmaps*** and ***scatter plots*** are good visualizations for this analysis.
##### Example Combinations:
- Selling_Price vs Original_Price
- Ratings vs Reviews

### Correlation Plot (Heatmap)
A heatmap helps visualize the correlation matrix between continuous variables. It shows the strength and direction (positive or negative) of the correlation.

```python
# Correlation matrix
corr_matrix = df[['Selling_Price', 'Original_Price', 'Discount_Percentage', 'Ratings']].corr()

# Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Continuous Variables')
plt.show()
```

### Scatter Plot
A scatter plot helps us visually inspect the relationship between two continuous variables.

```python
# Scatter plot between Selling_Price and Original_Price
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='Selling_Price', y='Original_Price', hue='Brand', palette='Set1',legend = False)
plt.title('Scatter Plot: Selling Price vs Original Price')
plt.xlabel('Selling Price')
plt.ylabel('Original Price')
plt.show()
```

### Categorical to Categorical Variables (Crosstab)
For categorical variables, we often want to understand how the categories of one variable relate to the categories of another. ***Crosstab*** is a great way to see the frequency distribution of variables.
##### Example Combinations:
- Brand vs OS
- Brand vs Ratings (although Ratings could also be treated as continuous, here it could be binned into categories)

### Crosstab
A crosstab helps compute the frequency of occurrences for each combination of two categorical variables.

```python
# Crosstab between 'Brand' and 'OS'
crosstab_data = pd.crosstab(df['Brand'], df['OS'])
print(crosstab_data)

# Crosstab between 'Brand' and 'Discount Percentage' (Binned into categories)
df['Discount_Binned'] = pd.cut(df['Discount_Percentage'], bins=5, labels=["Low", "Medium", "High", "Very High", "Extreme"])
crosstab_discount = pd.crosstab(df['Brand'], df['Discount_Binned'])
print(crosstab_discount)
```
